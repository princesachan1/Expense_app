```python
df['categories'].unique()
```

```python
category_map = {
    'Arts, Crafts & Sewing': 'Lifestyle & Fashion',
    'Clothing, Shoes & Jewelry': 'Lifestyle & Fashion',

    'Beauty': 'Health & Personal Care',
    'All Beauty': 'Health & Personal Care',
    'Health & Personal Care': 'Health & Personal Care',

    'Electronics': 'Electronics & Tech',
    'All Electronics': 'Electronics & Tech',
    'Cell Phones & Accessories': 'Electronics & Tech',
    'Office Products': 'Electronics & Tech',
    
    'Tools & Home Improvement': 'Home & Tools',
    'Patio, Lawn & Garden': 'Home & Tools',
    'Appliances': 'Home & Tools',
    'Industrial & Scientific': 'Home & Tools',
    
    'Baby Products': 'Baby Products',
    'Baby': 'Baby Products',
    
    'Sports & Outdoors': 'Leisure & Hobbies',
    'Toys & Games': 'Leisure & Hobbies',
    'Musical Instruments': 'Leisure & Hobbies'
}
df['categories'] = df['categories'].replace(category_map)
```

```python
df['categories'].nunique()
```

```python
df = df.replace(r'["()]', '', regex=True)
df = df.replace('-', ' ', regex=True)
df = df.replace('/', '', regex=True)
df = df.replace('  ', ' ', regex=True)
df['title'] = df['title'].str.replace(r'["\d]', '', regex=True)
df.head(5)
```

```python
df['title'] = df['title'].str.replace('&amp;', '&')
df['title'] = df['title'].str.replace('&quot;', '')
df['title'] = df['title'].str.replace(r'[^\w\s]', '', regex=True)
df['title'] = df['title'].replace('', pd.NA)
df = df.dropna(subset=['title', 'categories'])
df['title'] = df['title'].str.replace(r'([a-z])([A-Z])', r'\1 \2', regex=True)
df['title'] = df['title'].str.replace(r'\s+', ' ', regex=True)
df['title'] = df['title'].str.strip()


```

```python
df['title'] = df['title'].str.lower()
df['categories'] = df['categories'].str.lower()
```

```python
df.to_csv('clean_data.csv', index = False)
```
